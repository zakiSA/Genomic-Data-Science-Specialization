Week 3: Dynamic Programming to Implement Edit Distance

```python
def edDistRecursive(x, y):
    if len(x) == 0:
        return len(y)
    if len(y) == 0:
        return len(x)
    else:
        distHor = edDistRecursive(x[:-1], y) + 1
        distVert = edDistRecursive(x, y[:-1]) + 1
        if x[-1] == y[-1]:
            distDiag = edDistRecursive(x[:-1], y[:-1])
        else:
            distDiag = edDistRecursive(x[:-1], y[:-1]) + 1
    return min(distHor, distVert, distDiag)        
        
```

```python
def editDistance(x, y):
    # create matrix of zeros
    num_row = len(x) + 1
    num_col = len(y) + 1
    D = np.zeros((num_row,num_col))
    for i in range(len(x) + 1):
        D[i][0] = i         # fill the 1st row with ascending numbers from 0 to len(x+1)
    for i in range(len(y) + 1):
        D[0][i] = i         # fill tthe 1st column with ascending numbers from 0 to len(y+1)
    for i in range(1, len(x) + 1): # By row 
        for j in range(1, len(y) + 1): #By column
            #Compute the value that will go in this location or cell in D
            distHor = D[i][j-1] + 1 # If we are coming from the cell to the left
            distVert = D[i-1][j] + 1 # If we are coming form the cell on top 
            if x[i-1] == y[j-1]:         # If we are coming form the cell diagonal left  
                distDiag = D[i-1][j-1] # If the 
            else:
                distDiag = D[i-1][j-1] + 1
            D[i][j] = min(distHor, distVert, distDiag) # This willpopualte each cell with a value
            
    return D[-1][-1]  # Will return the value of the right bottom most cell which is the edit disance between two strings      
```
