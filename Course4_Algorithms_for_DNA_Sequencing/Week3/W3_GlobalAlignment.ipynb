```python
alphabet = ['A', 'C','G','T']
# Scoring Matrix: Penatlies A-A=0, A-C=4, A-G=2,gap=8, 
score = [[0,4,2,4,8],  # Penalties for 1st row
         [4,0,4,2,8],  # Penalties for 2nd row
         [2,4,0,4,8],  # Penalties for 3rd row
         [4,2,4,0,8],  # enalties for 4th row
         [8,8,8,8,8]]   # Last row of gaps  
```

```python
import numpy as np
```

```python
def globalAlignment(x, y):
    # create matrix of zeros
    num_row = len(x) + 1
    num_col = len(y) + 1
    D = np.zeros((num_row,num_col))
    
    #Initialize 1st row
    for i in range(1, len(x) + 1):
        D[i][0] = D[i-1][0] + score[alphabet.index(x[i-1])][-1]# fill the 1st row starting with 0 and then adding penalties from score matrix of the element to the left of it
                                                               # alphabet.index will give us the index of the character x that is a nmber between 1-4 and take the last value [-1] of the list/row since we are skipping that character  
    #Initialize 1st column
    for i in range(1, len(y) + 1):
        D[0][i] = D[0][i-1] + score[-1][alphabet.index(y[i-1])]# get character above it in col and get character in last row
                                        
    for i in range(1, len(x) + 1): # By row 
        for j in range(1, len(y) + 1): #By column
            #Compute the value that will go in this location or cell in D
            distHor = D[i][j-1] + score[-1][alphabet.index(y[j-1])] # Score of the last row. Penalty for skipping a character in y
            
            distVert = D[i-1][j] + score[alphabet.index(x[i-1])][-1] # Score of the last col.Penalty for skipping a character in x
            
            if x[i-1] == y[j-1]:         # If we are coming form the cell diagonal left  
                distDiag = D[i-1][j-1] # If the characters amtch
            else:
                distDiag = D[i-1][j-1] + score[alphabet.index(x[i-1])][alphabet.index(y[j-1])] # Get penalties if characters dont match
                
            D[i][j] = min(distHor, distVert, distDiag) # This willpopualte each cell with a value
            
    return D[-1][-1]  # Will return the value of the right bottom most cell which is the edit disance between two strings      
```

```python
# Identiacl strings
x = 'TACCAGATTCGA'
y = 'TACCAGATTCGA'
globalAlignment(x,y)    
```

```python
# 1 Skipped character in strings
x = 'TACCAGATTCGA'
y = 'TACCAAGATTCGA'
globalAlignment(x,y)    
```

```python
# 1 transersion, repalcement of G-C with character in strings
x = 'TACCAGATTCGA'
y = 'TACCACATTCGA'
globalAlignment(x,y)  
```

```python
# 1 transversion in the string G-A
x = 'TACCAGATTCGA'
y = 'TACCAAATTCGA'
globalAlignment(x,y)    
```
